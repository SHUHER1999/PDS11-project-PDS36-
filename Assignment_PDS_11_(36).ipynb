```python
import pandas as pd

# Load the dataset file
url = "https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

import pandas as pd
import matplotlib.pyplot as plt

# Filter passengers who were alive (assuming 'Survived' column exists, where 1 means alive)

df['Number of Relatives'] = df['SibSp'] + df['Parch']
alive_passengers = df[df['Survived'] == 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(alive_passengers['Number of Relatives'], alive_passengers['Age'], alpha=0.5)
plt.xlabel('Number of Relatives')
plt.ylabel('Age')
plt.title('Scatter Plot of Age vs Number of Relatives (Alive Passengers)')
plt.grid(True)
plt.show()
```
